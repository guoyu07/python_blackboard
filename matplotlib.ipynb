```python
from IPython.display import display
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import math
```

# 移动轴线

```python
def move_axis(y_axis_base = 0, x_axis_base=0):
    ax = plt.gca()
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    if(y_axis_base != None):
        ax.spines['bottom'].set_position(('data',y_axis_base))
    ax.yaxis.set_ticks_position('left')
    if(x_axis_base != None):
        ax.spines['left'].set_position(('data',x_axis_base))
```

# 正弦、余弦图

```python
plt.figure(1)
X = np.linspace(-np.pi, np.pi, 100, endpoint=True)
siny = np.sin(X)
cosy = np.cos(X)

#移动轴线
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

#设置刻度，使用了LATEX语法
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, +1], [r'$-1$', r'$0$', r'$+1$'])

#设置线条颜色
line_sin, = plt.plot(X, siny, color="blue", linewidth=1, linestyle="--")
line_cos, = plt.plot(X, cosy, color="red", linewidth=1, linestyle="-")

#添加图例
plt.legend([line_sin, line_cos],[r"sin(x)", r"cos(x)"], loc='upper left')
plt.title(u"正弦、余弦")

#用来正常显示中文标签  
plt.rcParams['font.sans-serif']=['Droid Sans Fallback']
#用来正常显示负号 
plt.rcParams['axes.unicode_minus']=False

#线上添加点
m = 2 * np.pi / 3
n = 2 * np.pi / 3
plt.scatter([m, ], [np.sin(m), ], 50, color='blue')
plt.scatter([n, ], [np.cos(n), ], 50, color='red')

#给m点添加注释，使用了LATEX语法
plt.annotate(r'$sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',xy=(m, np.sin(m)), xycoords='data',xytext=(+10, +30), 
    textcoords='offset points', fontsize=16, arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
#给n点添加注释，使用了LATEX语法
plt.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$',xy=(n, np.cos(n)), xycoords='data',xytext=(-90, -50), 
    textcoords='offset points', fontsize=16, arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

#设置标签字体大小和背景色
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor='Green', edgecolor='None', alpha=0.10))
plt.show()
```

# 3D

```python
from mpl_toolkits.mplot3d import Axes3D 
fig = plt.figure()  
ax = Axes3D(fig)  
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.15)  
X, Y = np.meshgrid(X, Y)  

R = np.sqrt(X**2 + Y**2)  
Z = np.sin(R)  
  
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.spring)  
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=plt.cm.winter)  
ax.set_zlim(-2,2)

ax.xaxis.set_label("X")
ax.yaxis.set_label("Y")
ax.zaxis.set_label("Z")

plt.show()  
```

# $f(x)=(1+\frac{1}{x})^x$

```python
fig = plt.figure(figsize=(16,12))
nums = 1000
X = np.linspace(-nums/2, nums/2, nums + 1, endpoint=True)
X1 = X[X > 0] #排除0
X2 = X[X < 0] #排除0
X2 = X2[X2 != -1] #排除-1，避免1+1/(-1)=0
Y1 = 1 + 1/X1
Y2 = 1 + 1/X2
Y12 = np.power(Y1,X1)
Y22 = np.power(Y2,X2)
Y3 = X*0 + np.e #f(x) = e

plt.plot(X1,Y12, color="blue")
plt.plot(X2,Y22, color="red")
plt.plot(X,Y3, color="green", linestyle="--")

#移动轴线
move_axis(None)

plt.show()

print Y12[nums/2 - 5:]
print Y22[nums/2 - 5:]
print np.e
```

# sigmoid

```python
def sigmoid(x):
    return 1/(1 + np.power(np.e, -1 * x))

fig = plt.figure()
nums = 12
X = np.linspace(-nums/2, nums/2, 100, endpoint=True)
Y = sigmoid(X)
plt.plot(X,Y)
x0, y0 = 0, sigmoid(0)
plt.scatter([x0, ], [y0, ], 50, color='red')
plt.annotate(r'$sigmoid(0)=\frac{1}{1 + e^{-x=0}}$',xy=(x0, y0), xycoords='data',xytext=(+30, -3), 
    textcoords='offset points', fontsize=16, arrowprops=dict(arrowstyle="->"))
move_axis(y_axis_base=None)
plt.show()
print sigmoid(0)
```
